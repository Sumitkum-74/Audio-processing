```python
#---------------------------------------------------- for audio processing -------------------------------------------
import librosa
import matplotlib.pyplot as plt

# Load the audio signal
audio_signal, sample_rate = librosa.load("song.mp3")

# Extract the first 10 seconds of the audio file
segment_1 = audio_signal[:10 * sample_rate]

# Extract the last 10 seconds of the audio file
segment_2 = audio_signal[-10 * sample_rate:]

# Perform Fourier Transform on the first segment
fft_segment_1 = np.fft.fft(segment_1)

# Perform Fourier Transform on the second segment
fft_segment_2 = np.fft.fft(segment_2)
# Compute the spectrogram
spectrogram = librosa.core.stft(audio_signal, n_fft=1050)

# Plot the waveform and spectrogram
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 5))

# Plot the waveform
ax1.plot(audio_signal)
ax1.set_xlabel("Time (seconds)")
ax1.set_ylabel("Amplitude")
ax1.set_title("Waveform")

# Plot the spectrogram
ax2.imshow(np.log(np.abs(spectrogram) ** 2), origin="lower")
ax2.set_xlabel("Time (seconds)")
ax2.set_ylabel("Frequency (Hz)")
ax2.set_title("Spectrogram")
plt.tight_layout()
plt.show()

#plot the fft of first segment
plt.plot(np.arange(len(fft_segment_1) // 2), np.abs(fft_segment_1[:len(fft_segment_1) // 2]))
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("FFT of the first segment")
plt.show()
#plot the fft of second segment
plt.plot(np.arange(len(fft_segment_2) // 2), np.abs(fft_segment_2[:len(fft_segment_2) // 2]))
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("FFT of the second segment")
plt.show()

# Plot the spectrogram of the first 5 seconds of the audio file
plt.imshow(np.log(np.abs(spectrogram[:5 * sample_rate]) ** 2), origin="lower")
plt.xlabel("Time (seconds)")
plt.ylabel("Frequency (Hz)")
plt.title("Spectrogram of the first 5 seconds")
plt.show()

# Plot the spectrogram of the middle 5 seconds of the audio file
plt.imshow(np.log(np.abs(spectrogram[len(audio_signal) // 2 - 5 * sample_rate:len(audio_signal) // 2 + 5 * sample_rate]) ** 2), origin="lower")
plt.xlabel("Time (seconds)")
plt.ylabel("Frequency (Hz)")
plt.title("Spectrogram of the middle 5 seconds")
plt.show()

# Plot the spectrogram of the last 5 seconds of the audio file
plt.imshow(np.log(np.abs(spectrogram[-5 * sample_rate:]) ** 2), origin="lower")
plt.xlabel("Time (seconds)")
plt.ylabel("Frequency (Hz)")
plt.title("Spectrogram of the last 5 seconds")
plt.show()

```
